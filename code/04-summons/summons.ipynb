```python
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv("summons.csv")
```

```python
df.head(50)
```

```python
df.plot()
```

```python
df['change'] = round((df['2016']-df['2009']) / df['2009']*100, 0)
```

```python
df2 = df[df['2009'] > 100]
```

```python
df2
```

```python
#df_top10 = df2.sort_values("Value", ascending=False).head(10)


df_top50 = df2.sort_values('change', inplace=True).head(50)
```

```python
df_top50
```

```python
df_top50.plot(x='Violation', y='change', kind='barh')
```

```python
df_top10 = df2.sort_values('change').head(10)
```

```python
df_top10
```

```python
ax = df_top10.plot(x='Violation', y='change', kind='barh')
ax.set_title("Top 10 reduction in Summons in NYC from 2009 to 2016, percentage", pad=10)
```

```python
df_top10incr = df2.sort_values('change', ascending=False).head(10)
```

```python
ax2 = df_top10incr.plot(x='Violation', y='change', kind='barh')
ax2.set_title("Top 10 increase in Summons in NYC from 2009 to 2016, percentage", pad=10)
```
