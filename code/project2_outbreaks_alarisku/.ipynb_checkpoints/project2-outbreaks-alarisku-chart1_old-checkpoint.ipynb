```python
df = pd.ExcelFile("NationalOutbreakPublicDataTool.xlsx").parse()
```

```python
#df.head()
```

```python
#Getting only cases with deaths
df = df[df['Deaths']>0]
```

```python
df.head()
```

```python
def booh(row):
    if row['Beef']:
        val = 'Meat'
    elif row['Fruits']:
        val = 'Vegetable'
    else:
        val = 'Other'
    return val
```

```python
df['Main Category'] = df.apply(booh, axis=1)
```

```python
df.loc[df['IFSAC Category'] == 'Pork', 'C'] = 'Meat'
```

```python
df.head()
```

```python
df['newcolumn'] = np.where(df['IFSAC Category'] == 'Pork', 'Meat', 'Other')
df['newcolumn'] = np.where(df['IFSAC Category'] == 'Beef', 'Meat', 'Other')
df.head(100)
```

```python
x = df.groupby('IFSAC Category').Deaths.sum()
x
```

```python
df_grouped = x.to_frame(name='deaths')
df_grouped.reset_index()
```
