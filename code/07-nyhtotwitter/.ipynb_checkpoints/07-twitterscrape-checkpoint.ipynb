```python
driver = webdriver.Chrome()
```

```python
driver.get("https://twitter.com/search?f=tweets&vertical=default&q=%23nyht%C3%B6kaura&src=typd")
```

```python
#dates =  driver.find_elements_by_class_name("js-short-timestamp")  
#for date in dates:
#    print(date.text)
```

```python
#This takes like 20 minutes
body = driver.find_element_by_tag_name("body")
time.sleep(1)

for _ in range(3000):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.2)
    
dates =  driver.find_elements_by_class_name("js-short-timestamp")    

rows = []

for date in dates:
    row = {}
    row['date'] = date.text
    rows.append(row)
```

```python
df = pd.DataFrame(rows)
df
```

```python
df.to_csv('nyhto_twitter.csv')
```

```python
#added the year 2018 in sublime and excel. created 2 separate files because the date was different format. Next I'll convert the dates to datetimeformat and then merge the to datarames back together
df = pd.read_csv("nyhto_twitter_edited.csv")
df2018 = pd.read_csv("nyhto_twitter_edited2018.csv")
```

```python
df.head()
```

```python
df2018['date'] = pd.to_datetime(df2018['date'], format="%b %d %Y")
```

```python
df['date'] = pd.to_datetime(df['date'], format="%d %b %Y")
```

```python
df.head()
```

```python
df2018.head()
```

```python
frames = [df, df2018]
df = pd.concat(frames)
df
```

```python
ax = df.resample('M', on="date").count().date.plot()
ax.grid(False, axis='x')
ax.xaxis.label.set_visible(False)
ax.set_yticks([100,200,600])
ax.set_yticklabels(["100","200","600 tweets\nin a month\nwith #nyhtökaura"])
ax.set_title('The Rise and Fall of the "Nyhtökaura" Hype in Twitter', pad=20, fontsize=13, loc='left')

```
