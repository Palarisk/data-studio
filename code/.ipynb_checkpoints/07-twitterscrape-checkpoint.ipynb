```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://twitter.com/search?f=tweets&vertical=default&q=%23nyht%C3%B6kaura&src=typd")
```

```python
#dates =  driver.find_elements_by_class_name("js-short-timestamp")  
#for date in dates:
#    print(date.text)
```

```python
body = driver.find_element_by_tag_name("body")
time.sleep(1)

for _ in range(3000):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.2)
    
dates =  driver.find_elements_by_class_name("js-short-timestamp")    

rows = []

for date in dates:
    row = {}
    row['date'] = date.text
    rows.append(row)
```

```python
df = pd.DataFrame(rows)
df
```

```python
df.to_csv('nyhto_twitter.csv')
```
